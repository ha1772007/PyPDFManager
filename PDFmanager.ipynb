```python
def ocr():
  try:
    import ocrmypdf
    print('ocrmypdf imported success: ')
    code = True
  except:
    print('installing ocrmypdf')
    !apt install ocrmypdf
    code = False
  ifmount = input('Enter if you want to mount GDrive or not [Y/N][Default : Y]').upper() != 'N'
  if(ifmount):
    try:
      from google.colab import drive
      print('\n Plz allow with all access if prompt for permission \n')
      drive.mount('/content/gdrive/')
    except:
      print('could not mount drive')
  else:
    print('Did not allowed To Mount')
  if (code):
    print('Now fill the following details')
    input_pathz = input('Enter the path file to ocr [Default = 1.pdf]') or '1.pdf'
    if(ifmount):
      input_path = '/content/gdrive/MyDrive/Colab Notebooks/src/'+input_pathz
    else:
      input_path = input_pathz
    print(input_path+ '\n')
    output_pathz = input('Enter the output path [Default = input_path_ocr.pdf]') or (input_pathz.split('.pdf')[0] + 'ocr.pdf')
    if(ifmount):
      output_path = '/content/gdrive/MyDrive/Colab Notebooks/src/'+output_pathz
    else:
      output_path = output_pathz
    print(output_path+ '\n')
    ocrmypdf.ocr(input_path, output_path,lang ='eng',redo_ocr=True)
    print('\n\n\n\nAll Process Ended\nThanks')
def pdftoimg():
  try:
    import fitz
    print('PyMuPDF imported sucessfully')
    status = True
  except:
    print('PyMuPDF could not be imported')
    status = False
    !pip install PyMuPDF
    print('PyMuPDF installation code executed success')
    pdftoimg()
  try:
    from tqdm import tqdm
  except:
    !pip install tqdm
    from tqdm import tqdm
  if(status):
    try:
      ifmount = input('Do you Want to Use Google Drive [Y/N][Defaut = Y]').upper() != 'N'
      from google.colab import drive
      drive.mount('/content/gdrive/')
      print('Drive Mounted')
    except:
      print('Could Not Perform Google Drive Mounting')
    input_pathz = input('Enter the path file to ocr [Default = 1.pdf]') or '1.pdf'
    if(ifmount):
      input_path = '/content/gdrive/MyDrive/Colab Notebooks/src/'+input_pathz
    else:
      input_path = input_pathz
    print(input_path+ '\n')
    output_pathz = input('Enter the output Folder path [Default = input_path_ocr/') or (input_pathz.split('.pdf')[0] + '/')
    if(ifmount):
      output_path = '/content/gdrive/MyDrive/Colab Notebooks/src/'+output_pathz
    else:
      output_path = output_pathz
    print(output_path+ '\n')
    !mkdir {output_path}
    pdf = fitz.open(input_path)
    try:
      zoom = int(input('input value of zoom[Default = 3]')) or 3
    except:
      zoom = 3
    for no, page in enumerate(tqdm(pdf, desc='converting PDF to img...')):
      mat = fitz.Matrix(zoom,zoom)
      savepath = output_path + str(no+1)+'.png'
      page.get_pixmap(matrix=mat).pil_save(savepath)
    print('/nTask Completed')

def ask():
  print('Enter\n1. --> OCR\n2. --> PDF to Img\n')
  todo = int(input('Enter What Do you want to do: '))
  if(todo == 1):
    ocr()
  elif(todo == 2):
    pdftoimg()
ask()


```
